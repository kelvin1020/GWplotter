```python
import json
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

def readpsd(detector):
    
    # 定义要读取的JSON文件的路径
    file_path = './data/detectors/' + detector + '.json'
    
    # 打开并读取JSON文件
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
    return np.array(data['data']).T
```

```python
DectorList = ['EPTA', 'IPTA', 'Gaia', 'LISA', 'Taiji', 'TianQin', 'aLIGODesign', 'aVirgo', 'KAGRA', 'ET']
```

```python
for i in range(len(DectorList)):
    runtext = DectorList[i] + ' = ' + 'readpsd("' + DectorList[i] + '")'
    exec(runtext)
```

```python
#绘制 频率 - sqrt(PSD) 图 

#绘图字体参数设置
matplotlib.rcParams.update({'font.size': 15, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})

plt.figure(figsize=(10,6))

for i in range(len(DectorList)):
    runtext = plottext = "plt.loglog(" + DectorList[i] + '[0], ' + DectorList[i] + '[1], '   + 'label = "' + DectorList[i] +'")'
    exec(plottext)




plt.xlim(1e-10, 1e6)
plt.ylim(1e-26, 1e-12)
plt.xlabel("Frequency / Hz")
plt.ylabel("Characteristic Strain")
# plt.ylabel("$\sqrt{S_{n}(f) \ / \ (Hz^{-1})}$")
plt.legend(loc = 'upper right')
plt.grid(linestyle = "dotted", color = "#d3d3d3" , which="both")
plt.tight_layout()
plt.show()
```

```python
#绘制 频率 - sqrt(PSD) 图 

#绘图字体参数设置
matplotlib.rcParams.update({'font.size': 15, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})

plt.figure(figsize=(10,6))

for i in range(len(DectorList)):
    runtext = plottext = "plt.loglog(" + DectorList[i] + '[0], ' + DectorList[i] + '[1] / np.sqrt(' +  DectorList[i] + '[0]), '   + 'label = "' + DectorList[i] +'")'
    exec(plottext)


plt.xlim(1e-10, 1e6)
plt.ylim(1e-26, 1e-6)
plt.xlabel("Frequency / Hz")
plt.ylabel("$(\mathrm{Power\,~Spectral\,~Density}~/~\mathrm{Hz}^{-1})^{1/2}$")
plt.legend(loc = 'upper right')
plt.grid(linestyle = "dotted", color = "#d3d3d3" , which="both")
plt.tight_layout()
plt.show()
```
