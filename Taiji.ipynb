```python
import numpy as np
```

```python
data =  np.loadtxt('taiji_3.txt').T
```

```python
data
```

```python
f = data[0][data[0] > 1e-5]
```

```python
s_n = data[1][data[0] > 1e-5]
```

```python
h_n = s_n  * np.sqrt(f)
```

```python
data2 = np.vstack([f, h_n]).T
```

```python
data3 = [[float(item[0]), float(item[1])] for item in data2]
data3
```

```python
data4 = {'data':data3}
```

```python
data4
```
